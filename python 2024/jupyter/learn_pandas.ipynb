```python
import pandas as pd
import matplotlib.pyplot as plt
import random

slope = 0
intercept = 0

learning_rate = 0.0001  


data = {'x': [random.randint(0, 100) for _ in range(250)],
        'y': [random.randint(0, 50) for _ in range(250)]}

df = pd.DataFrame(data, columns=['x', 'y'])

num_iterations = 100

for _ in range(num_iterations):
  
    slope_gradient = sum((slope * df['x'] + intercept - df['y']) * df['x']) / len(df['x'])
    intercept_gradient = sum(slope * df['x'] + intercept - df['y']) / len(df['x'])
    
 
    slope -= learning_rate * slope_gradient
    intercept -= learning_rate * intercept_gradient

    plt.scatter(df['x'], df['y'], color='blue')
    plt.plot(df['x'], slope * df['x'] + intercept, color='red')

plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['Fitted line', 'Data points'])
plt.show()
```
