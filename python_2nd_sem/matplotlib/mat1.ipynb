```python
 x = np.linspace(-5, 5, 100)
 y = np.sign(x)
 
 # Plotting
 plt.step(x, y, where='mid')
 plt.xlabel('x')
 plt.ylabel('sign(x)')
 plt.title('Step Function')
 plt.grid(True)
 plt.show()
```

```python
t = np.linspace(0, 2*np.pi, 100)
x = np.sin(t)
y = np.cos(t)

# Plotting
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Parametric Curve: x = sin(t), y = cos(t)')
plt.grid(True)
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 5, 100)
y = np.exp(-x)

# Plotting
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('e^(-x)')
plt.title('Exponential Decay')
plt.grid(True)
plt.show()
```
