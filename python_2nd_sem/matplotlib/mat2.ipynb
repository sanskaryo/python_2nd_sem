```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi,100)
y = np.sin(x)


plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sinusoidal Curve')
plt.grid(False)
plt.show()
```

```python
# Generate data
x = np.linspace(-5, 5, 100)
y = x**2

# Plotting
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('$x^2$')
plt.title('Parabolic Curve')
plt.grid(True)
plt.show()
```
