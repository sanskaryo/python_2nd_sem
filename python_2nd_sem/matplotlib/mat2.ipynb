```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,2*np.pi,100)
y = np.sin(x)


plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sinusoidal Curve')
plt.grid(False)
plt.show()
```

```python
import numpy as np

# Define the range of x values from 0 to 2*pi
x = np.linspace(0, 2 * np.pi, 100)

# Calculate the corresponding y values for sin(x), cos(x), and tan(x)
y_sin = np.sin(x)
y_cos = np.cos(x)
y_tan = np.tan(x)

# Plot the trigonometric functions
plt.plot(x, y_sin, label='sin(x)', color='blue')
plt.plot(x, y_cos, label='cos(x)', color='red')
plt.plot(x, y_tan, label='tan(x)', color='green')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trigonometric Functions')

# Set y-axis limits to avoid distortion caused by large tan(x) values
plt.ylim(-2, 2)

# Add a legend
plt.legend()

# Add a grid
plt.grid(True)

# Show the plot
plt.show()
```
